```python
# names of hurricanes
names = ['Cuba I', 'San Felipe II Okeechobee', 'Bahamas', 'Cuba II', 'CubaBrownsville', 'Tampico', 'Labor Day', 'New England', 'Carol', 'Janet', 'Carla', 'Hattie', 'Beulah', 'Camille', 'Edith', 'Anita', 'David', 'Allen', 'Gilbert', 'Hugo', 'Andrew', 'Mitch', 'Isabel', 'Ivan', 'Emily', 'Katrina', 'Rita', 'Wilma', 'Dean', 'Felix', 'Matthew', 'Irma', 'Maria', 'Michael']

# months of hurricanes
months = ['October', 'September', 'September', 'November', 'August', 'September', 'September', 'September', 'September', 'September', 'September', 'October', 'September', 'August', 'September', 'September', 'August', 'August', 'September', 'September', 'August', 'October', 'September', 'September', 'July', 'August', 'September', 'October', 'August', 'September', 'October', 'September', 'September', 'October']

# years of hurricanes
years = [1924, 1928, 1932, 1932, 1933, 1933, 1935, 1938, 1953, 1955, 1961, 1961, 1967, 1969, 1971, 1977, 1979, 1980, 1988, 1989, 1992, 1998, 2003, 2004, 2005, 2005, 2005, 2005, 2007, 2007, 2016, 2017, 2017, 2018]

# maximum sustained winds (mph) of hurricanes
max_sustained_winds = [165, 160, 160, 175, 160, 160, 185, 160, 160, 175, 175, 160, 160, 175, 160, 175, 175, 190, 185, 160, 175, 180, 165, 165, 160, 175, 180, 185, 175, 175, 165, 180, 175, 160]

# areas affected by each hurricane
areas_affected = [['Central America', 'Mexico', 'Cuba', 'Florida', 'The Bahamas'], ['Lesser Antilles', 'The Bahamas', 'United States East Coast', 'Atlantic Canada'], ['The Bahamas', 'Northeastern United States'], ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], ['The Bahamas', 'Cuba', 'Florida', 'Texas', 'Tamaulipas'], ['Jamaica', 'Yucatn Peninsula'], ['The Bahamas', 'Florida', 'Georgia', 'The Carolinas', 'Virginia'], ['Southeastern United States', 'Northeastern United States', 'Southwestern Quebec'], ['Bermuda', 'New England', 'Atlantic Canada'], ['Lesser Antilles', 'Central America'], ['Texas', 'Louisiana', 'Midwestern United States'], ['Central America'], ['The Caribbean', 'Mexico', 'Texas'], ['Cuba', 'United States Gulf Coast'], ['The Caribbean', 'Central America', 'Mexico', 'United States Gulf Coast'], ['Mexico'], ['The Caribbean', 'United States East coast'], ['The Caribbean', 'Yucatn Peninsula', 'Mexico', 'South Texas'], ['Jamaica', 'Venezuela', 'Central America', 'Hispaniola', 'Mexico'], ['The Caribbean', 'United States East Coast'], ['The Bahamas', 'Florida', 'United States Gulf Coast'], ['Central America', 'Yucatn Peninsula', 'South Florida'], ['Greater Antilles', 'Bahamas', 'Eastern United States', 'Ontario'], ['The Caribbean', 'Venezuela', 'United States Gulf Coast'], ['Windward Islands', 'Jamaica', 'Mexico', 'Texas'], ['Bahamas', 'United States Gulf Coast'], ['Cuba', 'United States Gulf Coast'], ['Greater Antilles', 'Central America', 'Florida'], ['The Caribbean', 'Central America'], ['Nicaragua', 'Honduras'], ['Antilles', 'Venezuela', 'Colombia', 'United States East Coast', 'Atlantic Canada'], ['Cape Verde', 'The Caribbean', 'British Virgin Islands', 'U.S. Virgin Islands', 'Cuba', 'Florida'], ['Lesser Antilles', 'Virgin Islands', 'Puerto Rico', 'Dominican Republic', 'Turks and Caicos Islands'], ['Central America', 'United States Gulf Coast (especially Florida Panhandle)']]

# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M', '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M', '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded', '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B', '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B', '91.6B', '25.1B']

# deaths for each hurricane
deaths = [90,4000,16,3103,179,184,408,682,5,1023,43,319,688,259,37,11,2068,269,318,107,65,19325,51,124,17,1836,125,87,45,133,603,138,3057,74]

# write your update damages function here:
# 1
# Update Recorded Damages
updated_damages = []
def update_damages(damages):
    for record in damages:
        if record[-1] == "M":
            updated_damages.append(float(record[:-1])*1000000)
        elif record[-1] == "B":
            updated_damages.append(float(record[:-1])*1000000000)
        else:
            updated_damages.append("Damages not recorded")
    return updated_damages

update_damages(damages)
print(updated_damages)



# write your construct hurricane dictionary function here:
hurricanes = {}
def create_dict(names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths):
    #Create individual dictionaries for each hurricane
    for i in range(len(names)):
        hurricane_dict = {}
        hurricane_dict["Name"] = names[i]
        hurricane_dict["Month"] = months[i]
        hurricane_dict["Year"] = years[i]
        hurricane_dict["Max Sustained Wind"] = max_sustained_winds[i]
        hurricane_dict["Areas Affected"] = areas_affected[i]
        hurricane_dict["Damage"] = updated_damages[i]
        hurricane_dict["Deaths"] = deaths[i]
        #Appends individual dictonary to overall one with name as key
        hurricanes[names[i]] = hurricane_dict
    return hurricanes

create_dict(names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths)
print(hurricanes)



# write your construct hurricane by year dictionary function here:

hurricane_years = {}
def year_canes(hurricanes, names):
    #Iterate through each individual hurricane
    for i in range(len(hurricanes.items())):
        # Store the year of each hurricane and its dictionary
        current_year = hurricanes.get(names[i])["Year"]
        current_cane = hurricanes.get(names[i])
        # If year does not exist as a key in hurricane_years then create new key with dictionary as value
        if hurricane_years.get(current_year, 0) == 0:
            hurricane_years[current_year] = [current_cane]
        # If year does exist as a key, then append relevant dictionary to list of dictionaries.
        else:
            hurricane_years[current_year].append(current_cane)
    return hurricane_years

year_canes(hurricanes, names)
print(hurricane_years)
        

# write your count affected areas function here:

area_count = {}
def count_affected_areas(areas_affected):
    # Flattens out areas_affected into single list
    total_areas = [area for cane in areas_affected for area in cane]
    # Empty list to popualte with unique entries
    list_of_regions = []
    for area in total_areas:
        # if area affected has not been added to unique list yet then append
        if area not in list_of_regions:
            list_of_regions.append(area)
        # if area does exist in unique list then skip and continue
        else:
            continue
    # Iterate through each unique region, count number of entries in flattened list and add as a value where the key is the unique region
    for area in list_of_regions:
        area_count[area] = total_areas.count(area)
    
    return area_count

count_affected_areas(areas_affected)
print(area_count)


# write your find most affected area function here:

def most_affected_area(area_count):
    #Lambda function to sort dictionary by values rather than keys, in descending order
    sorted_areas = sorted(area_count.items(), key = lambda x: x[1], reverse = True)
    return sorted_areas

sorted_areas = most_affected_area(area_count)
# Return the first entry of created tuples list, 0th is the name of the area, 1st is the number of times hit by cat 5 hurricanes
most_aff_area = sorted_areas[0][0]
num_of_times = sorted_areas[0][1]

print(most_aff_area)
print(num_of_times)

# write your greatest number of deaths function here:

most_lethal = "Cuba I"
max_deaths = 0

def most_lethal_hurricane(hurricanes, most_lethal, max_deaths):
    # Iterate through hurricanes dictionary and compare number of deaths
    for i in range(len(hurricanes.items())):
        # If the number of deaths is higher then overwrite value
        if hurricanes.get(names[i])["Deaths"] > max_deaths:
            max_deaths = hurricanes.get(names[i])["Deaths"]
            most_lethal = hurricanes.get(names[i])["Name"]
        #If it isn't higher then continue
        else:
            continue
    
    return (most_lethal, max_deaths)

            
most_lethal_cane = most_lethal_hurricane(hurricanes, most_lethal, max_deaths)     
print("The most lethal hurricane in the data was " + most_lethal_cane[0])
print(str(most_lethal_cane[1]) + " people died during the hurricane.")




# write your catgeorize by mortality function here:

mortality_scale = {0: 0, 1: 100, 2: 500, 3: 1000, 4: 10000}

hurricane_mortality = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
def mort_hurricanes(hurricanes, names, mortality_scale):
    #Iterate through each individual hurricane
    for i in range(len(hurricanes.items())):
        # Store the deaths caused by each hurricane and its dictionary
        current_deaths = hurricanes.get(names[i])["Deaths"]
        current_cane = hurricanes.get(names[i])
        # If statement to see which mortality category each hurricane falls into, and add it into relevant category 
        if current_deaths == 0:
            hurricane_mortality[0].append(current_cane)
        elif current_deaths <= 100:
            hurricane_mortality[1].append(current_cane)
        elif current_deaths <= 500:
            hurricane_mortality[2].append(current_cane)
        elif current_deaths <= 1000:
            hurricane_mortality[3].append(current_cane)
        elif current_deaths <= 10000:
            hurricane_mortality[4].append(current_cane)
        else:
            hurricane_mortality[5].append(current_cane)
    return hurricane_mortality

hurricanes_by_mortality = mort_hurricanes(hurricanes, names, mortality_scale)
print(hurricanes_by_mortality)




# write your greatest damage function here:

def highest_damage(hurricanes):
    """Find the highest damage inducing hurricane and its total cost."""
    max_damage_cane = 'Cuba I'
    max_damage = 0
    for cane in hurricanes:
        if hurricanes[cane]['Damage'] == "Damages not recorded":
            pass
        elif hurricanes[cane]['Damage'] > max_damage:
            max_damage_cane = cane
            max_damage = hurricanes[cane]['Damage']
    return max_damage_cane, max_damage

max_damage_cane, max_damage = highest_damage(hurricanes)
print("The Hurricane that caused the most damage was " + max_damage_cane)
print(str(max_damage) + " dollars of damage was caused.")



# write your catgeorize by damage function here:

damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}

hurricanes_damage_scale = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}

def damage_canes(hurricanes, names, damage_scale):
    #Iterate through each individual hurricane
    for i in range(len(hurricanes.items())):
        #Store the damage caused by each hurricane and it's dictionary
        current_damage = hurricanes.get(names[i])["Damage"]
        current_cane = hurricanes.get(names[i])
        #If statement to see which category each hurricane falls into, appends it into relevant category
        #Damages not recorded have been appended into category 0, should most likely be added to N/A category
        if current_damage == "Damages not recorded" or current_damage == 0:
            hurricanes_damage_scale[0].append(current_cane)
        elif current_damage <= 100000000:
            hurricanes_damage_scale[1].append(current_cane)
        elif current_damage <= 1000000000:
            hurricanes_damage_scale[2].append(current_cane)
        elif current_damage <= 10000000000:
            hurricanes_damage_scale[3].append(current_cane)
        elif current_damage <= 50000000000:
            hurricanes_damage_scale[4].append(current_cane)
        else:
            hurricanes_damage_scale[5].append(current_cane)
    
    return hurricanes_damage_scale

hurricanes_by_damage = damage_canes(hurricanes, names, damage_scale)
print(hurricanes_by_damage)
```
